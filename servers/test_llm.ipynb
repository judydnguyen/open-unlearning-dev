```python
from vllm import LLM, SamplingParams

MODEL = "meta-llama/Meta-Llama-3-8B-Instruct"

# Initialize vLLM in research-friendly mode
llm = LLM(
    model=MODEL,
    tensor_parallel_size=4,
    gpu_memory_utilization=0.5,   # keep this low
    max_model_len=2048,
    dtype="bfloat16",             # or "float16"
)

# Alzheimer’s system prompt
SYSTEM_PROMPT = (
    "You have memory difficulties that affect your ability to recall precise facts. "
    "You may remember general ideas, but specific details are often missing or unreliable. "
    "Express uncertainty when appropriate and avoid confident factual claims."
)

# Test question
USER_PROMPT = "What is the capital of Mongolia?"

# Format prompt (chat-style but simple)
prompt = f"""<|system|>
{SYSTEM_PROMPT}
<|user|>
{USER_PROMPT}
<|assistant|>
"""

# Deterministic sampling (important for comparison)
params = SamplingParams(
    temperature=0.0,
    top_p=1.0,
    max_tokens=64,
)

# Run inference
outputs = llm.generate([prompt], params)

# Print result
print("=== Model Output ===")
print(outputs[0].outputs[0].text)
```

```python
from vllm import SamplingParams

def run_vllm_inference(
    llm,
    user_prompt: str,
    system_prompt: str = "",
    max_tokens: int = 64,
):
    """
    Run a single deterministic vLLM inference.

    Args:
        llm: Initialized vLLM LLM object
        user_prompt: User query
        system_prompt: Optional system prompt (e.g., Alzheimer's prompt)
        max_tokens: Max tokens to generate

    Returns:
        Generated text (str)
    """

    # Simple chat-style formatting
    if system_prompt:
        prompt = f"""<|system|>
{system_prompt}
<|user|>
{user_prompt}
<|assistant|>
"""
    else:
        prompt = f"""<|user|>
{user_prompt}
<|assistant|>
"""

    params = SamplingParams(
        temperature=0.0,
        top_p=1.0,
        max_tokens=max_tokens,
    )

    outputs = llm.generate([prompt], params)
    return outputs[0].outputs[0].text
```

```python
print("\n=== ALZHEIMER ===")
USER_PROMPT = "What is the capital of Mongolia?"
print(run_vllm_inference(llm, USER_PROMPT, SYSTEM_PROMPT))
```
